```python
def toggle(odd=True):
    def inner(_):
        inner.value = not inner.value
        return inner.value
    inner.value = odd
    return inner

df = pd.read_csv(artwork_csv_path,
                 nrows=5,
                 # skiprows=5,
                 # usecols=['artist', 'title'],
                 usecols=toggle(True)
                 # index_col='id'
                 )
df
```
